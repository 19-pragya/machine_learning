```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,500)
c = np.random.randn(500)
y = x**3 + c*100 + 100
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x.shape
```

```python
y.shape
```

```python
lg1 = LinearRegression()
```

```python
#since our training data should be 2 dimensional

x1 = x.reshape(-1,1)
```

```python
x1.shape
```

```python
#checking the dimensions
x1.ndim
```

```python
lg1.fit(x1,y)
```

```python
pred1 = lg1.predict(x1)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y,pred1)
```

```python
print(score)
```

```python
#since r2 score is not close to one so our model os not working perfectly
```
